```python
#importing libraries 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sbn
import numpy as np
```

```python
id = pd.read_csv('./iris.data.csv')
id
```

# **EXPLORATORY DATA ANALYSIS (EDA)**

```python
id.hist(figsize =(5,10))
plt.tight_layout()
plt.show()
```

```python
#boxplot with seaborn
sbn.boxplot(x='petallength', y='sepallength', data=id)
plt.title('Petal-Sepal Lengths Distribution')
plt.show()
```

```python
#Scatterplot
sbn.scatterplot(x='sepalwidth', y='sepallength', hue="class", data=id)
plt.title('scatterplot for the three species')
plt.show()

```

```python
#heatmap
sbn.heatmap(id.corr(), annot=True, cmap="PuBu")
plt.title('correlation of the dataset')
plt.show()
```

```python
#barchart
sbn.countplot(x='class', data=id)
plt.title('Barchart')
plt.show()
```

```python
#pie chart
plt.pie(id["class"].value_counts(), labels = id["class"].unique())
plt.title("piechart")
plt.show()
```
